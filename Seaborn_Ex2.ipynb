```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('Pokemon.csv',index_col=0,encoding='unicode_escape')
df.info()
```

```python
df
```

```python
#2.1 scatterplot: sb.swarmplot (phan loai voi cac diem khong chong cheo)
plt.figure(figsize=(8,6))
sb.swarmplot(data=df, x='Attack',y='Defense')
plt.show()
```

```python
#2.2 scatterplot: sb.jointplot 
plt.figure(figsize=(8,6))
sb.set()
sb.jointplot(data=df, x='Attack',y='Defense',kind='reg',color='orange')
plt.show()
```

```python
#Ex2.3 #regression plot x='	Attack', y='Speed' #regplot
sb.regplot(data=df,x='Attack', y='Speed' )
plt.show()
```

```python
#Ex2.4 boxplot cho toan bo du lieu
sb.boxplot(data=df, x='Total')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
plt.figure(figsize=(16,12))
sb.boxplot(data=df, x='Type 1',y='HP')
plt.show()
```
